## **Variations to try**


1.   Use more padding to include the chin region

```
!cd Wav2Lip && python inference.py --checkpoint_path "/content/gdrive/MyDrive/Wav2lip/wav2lip.pth" --face "/content/gdrive/MyDrive/Wav2lip/opfinal.mp4" --audio "/content/gdrive/MyDrive/Wav2lip/output10.wav" --pads 0 20 0 0
```

```
!cd Wav2Lip && python inference.py --checkpoint_path "/content/gdrive/MyDrive/Wav2lip/wav2lip_gan.pth" --face "/content/gdrive/MyDrive/Wav2lip/opfinal.mp4" --audio "/content/gdrive/MyDrive/Wav2lip/output10.wav" --pads 0 20 0 0
```

2.   Use resize_factor to reduce the video resolution, as there is a change you might get better results for lower resolution videos. Why? Because the model was trained on low resolution faces.

```
!cd Wav2Lip && python inference.py --checkpoint_path "/content/gdrive/MyDrive/Wav2lip/wav2lip.pth" --face "/content/gdrive/MyDrive/Wav2lip/opfinal.mp4" --audio "/content/gdrive/MyDrive/Wav2lip/output10.wav" --resize_factor 2
```
